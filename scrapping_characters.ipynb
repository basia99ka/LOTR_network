```python
import pandas as pd
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import logging
import os
import time

```

## Code

```python

# Chrome WebDriver instance, create driver
driver = webdriver.Chrome()
driver.maximize_window()

# Go to page with characters
page_url = "https://lotr.fandom.com/wiki/Category:The_Lord_of_the_Rings_characters"
driver.get(page_url)

# Wait for the button (Accept cookies) to be clickable
accept_button = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.XPATH, '//div[text()="ZAAKCEPTUJ WSZYSTKO"]'))
)
# Click the button - Accept cookies
accept_button.click()

# Wait for the character elements to load
character_elems = WebDriverWait(driver, 10).until(
    EC.visibility_of_all_elements_located((By.CLASS_NAME, 'category-page__member-link'))
)

# Get all character 
characters_LOTR = []
character_elems = driver.find_elements(By.CLASS_NAME, 'category-page__member-link')
for elem in character_elems:
    characters_LOTR.append({'character': elem.text})

pd.DataFrame(characters_LOTR)



```
